```python
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from IPython.display import display, Image
import time

# Set up Firefox options for headless mode
firefox_options = Options()
firefox_options.add_argument("--headless")

# Initialize the Firefox driver with headless options
driver = webdriver.Firefox(options=firefox_options)

# Specify the URL to visit
url = 'https://www.winamax.fr/paris-sportifs/match/45484412'

# Navigate to the specified URL
driver.get(url)

# Wait for the page to load
time.sleep(4)

# Locate the button using its selector and click it
button = driver.find_element(By.CSS_SELECTOR, "#tarteaucitronAllDenied2")
button.click()

# Optionally, take a screenshot to verify the action (if needed)
screenshot = driver.get_screenshot_as_png()
display(Image(screenshot))

# Clean up by closing the browser
driver.quit()
```
